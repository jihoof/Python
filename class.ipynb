```python
l = []
count = 0

for _ in range(5):
    tmp = int(input("성적을 입력하시오"))
    l.append(tmp)

a = sum(l) / len(l)

print(f"성적 평균은 {a}입니다.")

for i in l:
    if i >= 80:
        count += 1
    else:
        pass

print(count)
```
