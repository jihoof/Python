```python
l = []
count = 0

for _ in range(5):
    tmp = int(input("성적을 입력하시오"))
    l.append(tmp)

a = sum(l) / len(l)

print(f"성적 평균은 {a}입니다.")

for i in l:
    if i >= 80:
        count += 1
    else:
        pass

print(count)
```

```python
ftem = float(input("화씨온도를 입력하시오: "))
ctem = (ftem - 32) * 5/9
print(ctem)
```

```python
price = {"soda" : 700, "cola" : 400, "milk" : 600}

buy = input("구매할 물건의 이름을 입력하시오: ")
im = int(input("투입할 돈의 양을 입력하시오: "))
productPrice = price[buy]

tmp = im - productPrice

won500 = tmp // 500 
won100 = tmp % 500 // 100
print(won500, won100) 
```
