```python
price = {"soda" : 700, "cola" : 400, "milk" : 600}

buy = input("구매할 물건의 이름을 입력하시오: ")
im = int(input("투입할 돈의 양을 입력하시오: "))
productPrice = price[buy]

tmp = im - productPrice

won500 = tmp // 500 
won100 = tmp % 500 // 100
print(won500, won100) 
```

```python
def merge_list(list1, list2):
    list1.extend(list2)
    list3 = sorted(list1)
    return list3

a = [0,1,2,3]
b = [4,5,6,7]
a, b = b, a

ml1 = merge_list(a, b)
print(ml1)
```

```python
def minmax(list):
    max1 = max(list)
    min1 = min(list)
    list.remove(max(list))
    list.remove(min(list))
    return max1, min1, list

l = [0,1,2,5]

print(minmax(l))
```

```python
def coin(money):
    fh = money // 500
    oh = money % 500 // 100
    fih = money % 100 // 50
    th = money  % 50 //10
    return fh, oh, fih, th

money = int(input())
print(coin(money))
```
