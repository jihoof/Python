```python
db = client["school"]
students = db["students"]

st_20 = list(students.find({

        "age": {"$gt": 20}

}))

AorB = list(students.find({"name": {"$in": ["Alice", "Bob"]}})) 
csstudent = list(students.find({

    "$and" : [

        {"age": {"$gt": 18}},
        {"major": "CS"}

    ]

}))

passed = list(students.find({

    "$nor": [

        {"age": {"$lt": 15}},
        {"grade": "C"}

    ]

})) 

cs_passed = list(students.find({

    "$and": [

        {"major": "CS"},
        {"grade": {"$in": ["A", "B"]}},
        

    ]

}))

print(json.dumps(clean_doc(st_20), indent=4))
print(json.dumps(clean_doc(AorB), indent=4))
print(json.dumps(clean_doc(csstudent), indent=4))
print(json.dumps(clean_doc(passed), indent=4))
print(json.dumps(clean_doc(cs_passed), indent=4))
```

```python
cs_ee_good = list(students.find({

    "$and": [

        {"score": {"$gt": 80}},
        {"major": {"$in": ["CS","EE"]}}

    ]

}))

print(json.dumps(clean_doc(cs_ee_good), indent=1))
```

```python
#5. major 가 ME, CE, EE 중 하나이면서 grade가 "B"인 학생 찾기.

meceeeb = list(students.find({

    "$and": [

        {"major": {"$in": ["ME", "EE", "CE"]}},
        {"grade": "B"}

    ]

}))

print(meceeeb)
```

```python
#문요준 9:00 PM6. grade가 A 또는 B이지만, CS 전공이 아닌 학생 모두 찾기.

a = list(students.find({

    "$and": [

        {"grade": {"$in": ["A","B"]}},
        {"major": {"$nin": ["CS"]}}

    ]

}))

print(a)
```

```python
#7. 나이가 18보다 크고, (성적이 A이거나 점수가 85 이상) 인 학생 찾기.

b = list(students.find({

    "$and": [

        {"age": {"$gt": 18}},
        {"$or": [{"grade": "A"}, {"score": {"$gte": "85"}}]}
    ] 

}))

print_(b)
```
