```python
# grade가 A이고 score가 90 이상인 학생에게
# devices 필드를 빈 리스트([]) 로 새로 생성하라.
students.update_many({

    'grade' : 'A',
    'score' : {'$gte':90}

}, {

    '$set' : {

        'devices' : []

    }

})
```

```python
#age가 18 이상이고 major가 CS 또는 EE인 학생에게
#scores 필드를 [70, 80, 90] 으로 생성하라.

students.update_many({

    'major' : {'$in' : ["CS", "EE"]},
    'age' : {'$gte' : 18}

}, {

    '$set' : {

        'scores' : [70, 80, 90]

    }

})
```

```python
# (score가 85 이상이고 grade가 A) 또는
# (age가 20 이상이고 major가 CS)
# 인 학생에게:
# devices 필드 생성 → ["iPhone"]
# score를 2 증가

students.update_many({

    '$or' : [

        {'grade':'A','score':{'$gte':85}},
        {'age':{'$gte':20}, 'major':'CS'}

    ]

}, {

    '$set': {

        'devices' : [

            'iPhone'

        ]

    },

    "$inc" : {

        'score' : 2

    }

})
```

```python
# major가 EE이고 grade가 C가 아닌 학생에게
# scores 필드를 생성하고 [60, 70, 80] 을 넣어라.
# 그 후 score를 5 증가시켜라.

students.update_many({

    'major' : {'$ne' : "EE"},
    'grade' : {'$ne' : "C"}

}, {

    '$set' : {

        'scores' : [60, 70, 80],
        
    },
    
    '$inc' : {
        
        'score' : 5
            
            }

})
```

```python
# 다음 조건을 모두 만족하는 학생을 업데이트하라:
# age ≥ 18
# score ≥ 80
# (grade = A OR major = CS)
# grade ≠ C
# 업데이트:
# devices → ["Initialized"]
# scores → [70, 80]
# score +5

students.update_many({

    'age' : {'$gte' : 18},
    'score' : {'$gte' : 80},
    '$or' : [

        {'grade' : 'A'},
        {'major' : 'CS'}

    ],
    'grade' : {'$ne':'C'}

}, {

    '$set' : {

        'devices' : ['Initialized'],
        'scores' : [70, 80],
        
    },

    '$inc' : {

        'score' : 5
        

    }

})
```

```python
# grade가 A이고 score가 90 이상인 학생에게
# devices 배열을 생성하고 "iPhone" 을 추가

students.update_many({

    'grade' : 'A',
    'score' : {'$gte' : 90}

}, {

    '$push' : {

        'devices' : 'iPhone'

    }

})
```

```python
# age가 18 이상이고 major가 CS 또는 EE인 학생에게
# scores 배열을 생성하고 [80, 90] 을 한 번에 추가

students.update_many({

    'age' : {'$gte':18},
    'major' : {'$in':["CS","EE"]}

}, {

    '$push' : {

        'scores' : {'$each':[80,90]}

    } 

})
```

```python
# devices 배열이 없거나 이미 있든 상관없이,
# grade가 A인 학생에게 "iPad" 를 중복 없이 추가

students.update_many({

    'grade' : 'A'

},{

    '$addToSet' : {

        'devices' : "iPad"
        ""

    }

})
```

```python
# major가 CS이고 score가 85 이상인 학생에게
# devices 배열에 "MacBook", "iPhone" 을 중복 없이 한 번에 추가

students.update_many({

    'major' : "CS",
    'score' : {'$gte' : 85}

}, {

    '$addToSet' : {

        'devices' : {'$each' : ['MacBook', "iPhone"]}

    }

})
```

```python
# grade가 C인 학생 중
# scores 배열이 있다면,
# scores 배열에서 70 미만 점수를 제거

students.update_many({

    'grade' : 'C',
    'scores' : {'$exists' : True} 

}, {

    '$pull' : {

        'scores' : {'$lt' : 70}

    }

})
```

```python
# major가 EE이고 score가 80 미만인 학생의
# devices 배열에서 "Android" 를 제거하라.

students.update_many({

    'major' : 'EE',
    'score' : {'$lt' : 80}

}, {

    '$pull' : {

        'devices' : 'Android'
 
    }

})
```
